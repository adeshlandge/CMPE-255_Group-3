```python
#separate mal n female names
data_male = baby_data[baby_data['Gender']=='M']
data_female = baby_data[baby_data['Gender']=='F']
```

```python
data_female_pivot = data_female.pivot_table('Count', index = 'Year', columns = 'Name', aggfunc = sum)
data_male_pivot = data_male.pivot_table('Count', index = 'Year', columns = 'Name', aggfunc = sum)
# data_female_pivot = data_female.pivot_table('Count', columns = 'Name', aggfunc = sum)
# data_male_pivot = data_male.pivot_table('Count',  columns = 'Name', aggfunc = sum)
```

```python
data_grouped_female_list = data_grouped_female.values.tolist()
data_grouped_male_list = data_grouped_male.values.tolist()
```

```python
data_female_array = np.array(data_grouped_female_list)
data_female_array
data_male_array = np.array(data_grouped_male_list)
data_male_array
```
